```python
targetSite = 'https://music.bugs.co.kr/chart'
```

```python
request = requests.get(targetSite)
# print(request)
html = request.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
# print(soup)
```

```python
# 노래 제목 크롤링
titles = soup.findAll('p', {'class': 'title'})
# print(titles)
# print(len(titles))

for title in titles:
    # print(title.text.strip())
    # print(title.text.split('\n')[1])
    print(title.text.strip().split('\n')[0])
```

```python
# 아티스트 크롤링
artists = soup.findAll('p', {'class': 'artist'})
# print(artists)
# print(len(artists))

for artist in artists:
    # print(artist.text.strip())
    # print(artist.text.split('\n'))
    print(artist.text.strip().split('\n')[0])
```

```python
print(dt.now())
for i in range(len(titles)):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    print('{0:3d}위 {1} - {2}'.format(i + 1, artist, title))
```

```python
# 크롤링한 결과를 텍스트 파일에 저장한다.
# 크롤링한 데이터를 텍스트 파일에 저장하려는 경우 아래와 같이 파일을 열면 한글 인코딩 문제로 인해
# UnicodeEncodeError가 발생될 수 있다.
file = open('bugsTOP100.txt', 'w')
# UnicodeEncodeError: 'cp949' codec can't encode character '\xf1' in position 29: illegal multibyte 
# sequence 에러가 발생되면 아래와 같이 파일을 open 시키면 된다.
# file = open('bugsTOP100.txt', 'w', -1, 'UTF-8') 
file.write('{} 현재 Bugd 뮤직 실시간 TOP 100\n'.format(dt.now()))
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    data = '{0:3d}위 {1} - {2}'.format(i + 1, artist, title)
    file.write(data + '\n')
file.close()
print('bugsTOP100.txt 파일로 쓰기 완료')
```

```python
# with 구문을 사용해서 파일을 open하면 with 내부의 내용이 모두 실행되고 난 후 자동을 파일을 close 한다.
with open('bugsTOP100.txt', 'w') as file:
    file.write('{} 현재 Bugd 뮤직 실시간 TOP 100\n'.format(dt.now()))
    for i in range(100):
        artist = artists[i].text.strip().split('\n')[0]
        title = titles[i].text.strip().split('\n')[0]
        data = '{0:3d}위 {1} - {2}'.format(i + 1, artist, title)
        file.write(data + '\n')
print('bugsTOP100.txt 파일로 쓰기 완료')
```

```python
# 텍스트 파일에 저장된 데이터를 읽어서 화면에 출력한다.
try:
    file = open('bugsTOP100.txt', 'r')
    # readlines() 메소드는 텍스트 파일의 내용 전체를 한꺼번에 읽어 list로 변환한다.
    lines = file.readlines()
    # print(type(lines))
    for line in lines:
        # print(line, end = '')
        print(line.strip())
    file.close()
    print('bugsTOP100.txt 파일에서 읽기 완료')
except:
    print('디스크에 읽어올 파일이 없습니다.')
```

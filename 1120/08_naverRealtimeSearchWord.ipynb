```python
request = requests.get(targetSite, headers = header)
words = request.text
print(type(words))
print(words)
```

```python
# json       json  import .
import json
```

```python
# json  loads()   json  ()  .
searchWords = json.loads(words)
print(type(searchWords))
print(searchWords)
```

```python
#    .
ranks = searchWords['data']
print(type(ranks))
print(ranks)
```

```python
#            .
for rank in ranks:
    print('{0:2d} => {1} ({2})'.format(rank['rank'], rank['keyword'], rank['keyword_synonyms']))
```
